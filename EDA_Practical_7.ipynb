```python
# Odinal Encoding

import pandas as pd
import datetime
```

```python
df_base = datetime.datetime.today()
df_base
```

```python
df_date_list = [df_base - datetime.timedelta(days=x) for x in range(0,20)]
df_date_list
```

```python
df = pd.DataFrame(df_date_list)
```

```python
df.columns=['Day']
```

```python
df.head()
```

```python
df['day_of_week'] = df['Day'].dt.day_name()
```
