```python
cat_feature_with_na = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes =='O']

print(len(cat_feature_with_na))

for feature in cat_feature_with_na:
    print("{}:{}% missing values".format(feature,np.round(df[feature].isnull().mean()*100,3)))

```

```python
# Replace missing value with a new a label 

def replace_cat_feature(df,cat_feature_with_na):
    data = df.copy()
    data[cat_feature_with_na] = data[cat_feature_with_na].fillna('Missing')
    return data

df = replace_cat_feature(df,cat_feature_with_na)

df[cat_feature_with_na].isnull().sum()
```

```python
df.head()
```

```python
# Numerical variables which have missing values

numerical_feature_with_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

print(len(numerical_feature_with_nan))

for feature in numerical_feature_with_nan:
    print("{}:{}% missing values".format(feature,np.round(df[feature].isnull().mean()*100,3)))
```

```python
# Replacing the numerical missing values

for feature in numerical_feature_with_nan:
    # since there are outliers replacing nan with median
    
    median_value = df[feature].median()
    
    # create a new feature to capture nan values
    
    df[feature +' nan'] = np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value, inplace= True)
    
df[numerical_feature_with_nan].isnull().sum()
```

```python
df.head()
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature] = df['YrSold']-df[feature]
```
